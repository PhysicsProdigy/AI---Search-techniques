This will generate a file of Station like stops from the file stops.txt


```python
finalData = []
with open('stops.txt') as f:
  dataToArrays = []
  for line in f.readlines():
    fixedLine = []
    lss = line.split('","')
    for item in lss:
      itemstrip = item.replace('"','')
      if itemstrip != '\n':
        fixedLine.append(itemstrip)
      else:
        fixedLine.append("")
        dataToArrays.append(fixedLine)
    

    for data in dataToArrays:
      if data[2].endswith(" Station") and "opp" not in data[2] and "Quarantine" not in data[2] and "Power" not in data[2] and "Service" not in data[2] and "Fire" not in data[2] and "Police" not in data[2] and "near" not in data[2] and "Rd" not in data[2] and "Ambulance" not in data[2] and data not in finalData:
          print(data)
          finalData.append(data)
  f.close
```

This will read the file we made and compare the stations that exist according to wiki

```python
dictionary = {}
with open('Stations.txt',"w") as f:
  for data in finalData:
    i = 0
    for station in Stations:
      i += 1
      if data[2] == station + " Station" and data[2] not in dictionary:
        dictionary[data[2]] = i
        f.write(f"{data[2]},{data[3]},{data[4]}\n")
  #Adding the missing stations and their details
  f.write("Circular Quay,-33.8614,151.211\n")
  f.write("International Airport,-33.934058, 151.165783\n")
  f.write("Domestic Airport,-33.933653, 151.180708\n")
print(len(dictionary))
```

This Function will be called over and over again for users to manually input the time between two stations


```python
def writeCell(prevValue):
  if (prevValue):
    First_Station = prevValue
    Second_Station = input("Whats the next station?")
    if Second_Station == "Done":
      return 1,1,1,1,1 #Everything is finished and needs saving
    if Second_Station == "0":
      return 1,1,1,1,0 #Only the Route is finished, but start a new one
    print(f"Second Station is : {Second_Station}")

    First_Station_Index = -1
    Second_Station_Index = -1
    try:
        Time_Between = int(input("Whats the time between?"))
    except:
        print("You failed to write an integer... try again")
        return 0,0,0,0,0 #This will make the user try again
    for i in range(len(Stations)):
      if Stations[i] == First_Station:
        First_Station_Index = i
      if Stations[i] == Second_Station:
        Second_Station_Index = i
    if First_Station_Index == -1 or Second_Station_Index == -1:
        print(f"First station was {First_Station_Index} and second station was {Second_Station_Index}. If these are -1, they are not in the station list.")
        return 0,0,0,0,0 # This will make the user try again, mention which stations may haps been falsly spelt. Or are not in this list. 
    return First_Station_Index,Second_Station_Index,First_Station,Second_Station,Time_Between
  else:
    First_Station = input("Whats the first station?")
    print(f"First Station is : {First_Station}")
    return writeCell(First_Station)
```

This is what we ran to update the excel file so that we can update the values


```python
import pandas as pd
import openpyxl
import xlwt

#Initialises a blank DF
data = []
axis = []
for i in range(len(Stations)):
  axis.append(Stations[i])
  data.append([0] * len(Stations))
print(len(data),len(axis))

#Stores the first input as default value between 2 NODES
previousStation = ""
notConnectedStations = []

#Tries to read existing file and merges the values to the df
try:
  dataframe = pd.read_excel('TrainPairs.xlsx', sheet_name = 'edges')
  for i in range(len(data)):
    Connected = False
    for j in range(len(data)):
      data[i][j] = dataframe.values[i][j+1]
      if (dataframe.values[i][j+1]>0):
        Connected = True
    if Connected == False:
      notConnectedStations.append(dataframe.values[i][0])
    Connected = False
  print(f"Stations are not connected: {notConnectedStations}")
  
except:
  print('No file')

#Requests User manual input to link two nodes and mention the time between the 2 stations
while True:
  First_Station_Index,Second_Station_Index,First_Station,Second_Station,Time_Between = writeCell(previousStation)
  if First_Station == 0 and Second_Station == 0 and Time_Between == 0:
    continue 
  if First_Station == 1 and Second_Station == 1 and Time_Between == 1:
    break # Got to save the file some how
  if First_Station == 1 and Second_Station == 1 and Time_Between == 0:
    previousStation = ""
    continue # Route has ended but don't want to stop the loop. 
  if type(Time_Between) is int:
    data[First_Station_Index][Second_Station_Index] = Time_Between
    data[Second_Station_Index][First_Station_Index] = Time_Between
    previousStation = Second_Station
    print(f"From {First_Station} to {Second_Station} is {Time_Between} minutes.")
  else:
    print(f"Sorry you input a {type(Time_BB3etween)}")
    continue

#Creates the DF
df = pd.DataFrame(data,index = axis, columns = axis)

#Saves the file
print("Saving File")
df.to_excel('TrainPairs.xlsx', sheet_name='edges')
```
